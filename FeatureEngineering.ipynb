# Setup

```python
!pip install librosa matplotlib
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/Main_kepler_star_KIC12268220C.mp3'
y, sr = librosa.load(file_path)
```

# Display a Spectrogram

```python
plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```

# Compute a Mel-scaled Spectrogram

```python
# Compute the Short-Time Fourier Transform (STFT)
D = np.abs(librosa.stft(y))

# Compute the Mel-scaled Spectrogram from the STFT
mel_spectrogram = librosa.feature.melspectrogram(S=D**2, sr=sr, n_mels=128)

# Display the Mel-scaled Spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(mel_spectrogram, ref=np.max), sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-scaled Spectrogram')
plt.tight_layout()
plt.show()
```

# Short-Time Fourier Transform (STFT)

```python
stft = librosa.stft(y)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(stft)), sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Short-Time Fourier Transform')
plt.tight_layout()
plt.show()
```

# Beat Tracker

```python
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
print(f'Tempo: {tempo} BPM')
print(f'Beats: {beats}')
```

# Compute the Constant-Q Transform (CQT)

```python
cqt = librosa.cqt(y, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(cqt)), sr=sr, x_axis='time', y_axis='cqt_hz')
plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q Transform')
plt.tight_layout()
plt.show()
```
